Concept of Abstract Data Types
- Unlike different built-in data types that are limited to basic operations like addition, subtraction, division and multiplication, abstract data types perform specific requirement operations that are usually custom.
- Abstract data types allow for development of programs that can perform unique operations that are out of the range of the basic operations since it does not specify how data will be organized in memory and what algorithms will be used for implementing the operations.

Difference between ADT and Data structures
- ADTs are used to design and conceptualize data structures while Data Structures are used to implement data structures that understand the abstract concepts defined by ADTs.
- ADTs do not specify how operations are implemented or how data is structured while Data structures specify how to create and organize data types to implement the structure.
- ADTs provide a high level conceptual view of data structures and functionalities while Data structures while Data Structures model specific data structures based on primitive types.

Examples of ADTs 
- List ADT: This is implemented using using either an array or a linked list
- Stack ADT: This is implemented using either an array or a linked list following the stack concept.
- Queue ADT: This is implemented using either a linked list or an array following the First in First Out technique

- Structures
- Classes
- Enumerations
- Records

Arrays
- Array is a linear data structure where all elements that are of the same data type are arranged sequentially such that loacting elements is easier using their position in the array.
- In an array, memory is allocated to the whole array. 

```python
array = [[1,1],
         [2,2],
         [3,3],
         [4,4]]
print(array)
```

Linked Lists
- Linked lists unlike arrays allow easier and more efficient insertion and deletion operations. 
- In a list memory is allocated one by one to individual elements 

```python
linked_list= {
    0: (1,1),
    1: (2,2),
    2: (3,-1)
}


def display_linked_list():
    index=0
    while index != -1:
        value, next_index = linked_list[index]
        print(value, end=' -> ')
        index=next_index
    print('none')

print('linked_list')
display_linked_list()
```

Stacks
- A stack is a linear data structure that follows a particular order in which the operations are performed. 
- There's basically to main orders which include First In Last Out (FILO) and Last In First Out (LIFO)
- FILO, the element that is inserted first comes out last
- LIFO, the element that is inserted last comes out first

```python
stack = []
stack.append(1)
stack.append(2)
stack.append(3)
print('stack',stack)
```

Queues
- Queues are data structures that follow the First in, First out principle (FIFO) where the first person to be added to a queue is the first to be removed.

```python
from collections import deque

queue = deque()

queue.append(1)
queue.append(2)
queue.append(3)

print('queue', queue)
print('queue is now a list', list(queue))
```

Recursions
- This is a method through which problems and computational assignments are broken down into smaller sub-problems to find a solution. 

```python
def factorial(number):

    if number == 0:
        return 1
    else:
        return number * factorial(number-1)
    
print(factorial(6))
```

Trees
- A tree is a hierarchical non-linear data structure used to represent and organize data in which the top most node is called a root node and th nodes below it called a child node

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.children = []

    def add(self, *nodes):
        self.children.extend(nodes)

    def show(self, level=0):
        print("  " * level + str(self.value))
        for child in self.children:
            child.show(level + 1)

# COMPANY HIERARCHY
ceo = TreeNode("CEO")
cto = TreeNode("CTO")
cfo = TreeNode("CFO")
dev = TreeNode("Developer")
acct = TreeNode("Accountant")

ceo.add(cto, cfo)
cto.add(dev)
cfo.add(acct)

ceo.show()
```

An algorithm is the step by step sequence for solving a particular problem.
- Algorithms are used to automate decisions
- Algorihms are also used to make predictions through analyzing data 

Pseudocode for summation of numbers in an array

```python
'''

BEGIN 

    SET sum = 0
    FOR each element in array
        ADD element to sum
    END FOR
    PRINT SUM
END


'''
```

Pseudocode for matrix multiplication algorithm

```python
'''
BEGIN

Algorithm MatrixMultiplication(A, B)
    Input: 
        A is an m x n matrix
        B is an n x p matrix
    Output:
        C is an m x p matrix (result of A * B)

    1. Initialize matrix C as an m x p matrix with all elements set to 0
    2. For i = 1 to m:
        3. For j = 1 to p:
            4. C[i][j] = 0
            5. For k = 1 to n:
                6. C[i][j] = C[i][j] + A[i][k] * B[k][j]
    7. Return matrix C

END

'''
```

Different types of algorithms with short notes
Brute Force Algorithm
- This is a simple search strategy that systematiclaly explores every option until the problem's solution is discovered.
- It's usually employed when the problem to be solve is small anough to make an in-depth investigation.

Recursive Algorithm
- This is an algorithm that breaks down huge tasks into smaller similar problems that are quite easier to solve.

Divide and Conquer Algorithm
- This is an algorithm that involves breaking the given problem into smaller non-overlapping problems, solve them and use the solution of the small problems to find the overall result.

Dynamic Programming Algorithm
- This is an algortihm that optimizes recursive solutions when similar subproblems are called again.
- This stores solutions to subproblems so that each subproblem is solved only once.

Greedy Algorithm
- This is an algorithm that chooses the best option at each step to find a solution for the entire problem.

Backtracking Algorithm
- This algorithm explores all possible solutions to a problem while eliminating those that don't work.
